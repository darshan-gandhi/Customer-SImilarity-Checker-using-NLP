```python
mydict = dict(zip(preprocess.ENGL_CUST_NAME, preprocess.CUST_NAME))
```

```python
data1_custname=[]
for j in data1.CUST_NAME:
    i=j.replace(","," ")
    k=i.replace("."," ")
    l=k.replace("  "," ")
#     print(l)
    l_new=list(l.split(" "))
    for val in range(len(l_new)):
#         print(l_new[val])
        if l_new[val] in mydict:
#             print(l_new[val])
            l_new[val]=mydict[l_new[val]]
            
    store_val=" ".join(l_new)
    data1_custname.append(store_val)
#     print(store_val)
   
```

```python
# print(len(data1_custname))

data1['CUST_NAME']=data1_custname
data1
```

```python
data2_custname=[]
for j in data2.ENGL_CUST_NAME:
    i=j.replace(","," ")
    k=i.replace("."," ")
    l=k.replace("  "," ")
#     print(l)
    l_new=list(l.split(" "))
    for val in range(len(l_new)):
#         print(l_new[val])
        if l_new[val] in mydict:
#             print(l_new[val])
            l_new[val]=mydict[l_new[val]]
            
    store_val=" ".join(l_new)
    data2_custname.append(store_val)
```
