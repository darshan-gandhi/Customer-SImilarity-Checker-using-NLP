```python
newdata=pd.read_csv("output.csv")
newdata['LINE_NO']=newdata['LINE_NO'].astype(str)
newdata.head()
```

```python
from fuzzywuzzy import fuzz
```

```python
save_val=[]

val1=list(newdata['CUST_NAME'])
val2=list(newdata['ENGL_CUST_NAME'])

for i in range(len(val1)):

    Ratio = fuzz.token_sort_ratio(val1[i].lower(),val2[i].lower())
    save_val.append(Ratio)
```

```python
newdata['Ratio']=save_val
```
