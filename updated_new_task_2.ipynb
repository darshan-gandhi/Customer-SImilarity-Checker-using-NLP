```python
from fuzzywuzzy import fuzz
```

```python
val=list(ans['ENGL_CUST_NAME'])



for i in range(len(val)):
    val[i]=val[i].lower()
    
```

```python
for i in range(len(data_cust_acc)):
    data_cust_acc[i]=data_cust_acc[i].lower()
    
```

```python
from tqdm import tqdm
save_ratio=[]
save_cust=[]
final_ratio=[]
final_cust=[]
count=0

for i in tqdm(range(len(val))):
    save_ratio=[]
    save_cust=[]
    for j in data_cust_acc:
#         print(val[i],j)
        Ratio = fuzz.token_sort_ratio(val[i],j)
        if Ratio>=80:
            save_ratio.append(Ratio)
            save_cust.append(j)
        
    
    final_ratio.append(save_ratio)
    final_cust.append(save_cust)
```

```python
len(final_ratio)
len(final_cust)
```

```python
ans['Names']=final_cust
ans['Ratios']=final_ratio
```
