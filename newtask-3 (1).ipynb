```python
from fuzzywuzzy import fuzz
```

```python
from tqdm import tqdm
save_ratio=[]
save_cust=[]
final_ratio=[]
final_cust=[]
count=0
save_both=[]
for i in tqdm(range(len(val))):
    save_ratio=[]
    save_cust=[]
    save_both=[]
    for j in data_cust_acc:
#         print(val[i],j)
        Ratio = fuzz.token_sort_ratio(val[i],j)

        
        if Ratio>=80:
            inter_save=str(Ratio)+"|"+j
#             print(inter_save)
            new_save=list(inter_save.split("|"))
#             print(new_save)
#             save_ratio.append(Ratio)
#             save_cust.append(j)
            save_both.append(new_save)
#             count+=1
#             valcheck=max(int(save_both[]))
#             print(valcheck)
        
    if save_both==[]:
        final_ratio.append(None)
        final_cust.append(None)
    elif len(save_both)>0:
        
        
        valnext1=[int(row[0]) for row in save_both]
#         print(valnext1)
        save=valnext1.index(max(valnext1))
#         print(save)
        ans1=save_both[save]
    
        final_ratio.append(ans1[0])
        final_cust.append(ans1[1])
```
